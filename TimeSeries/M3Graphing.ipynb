```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
construction = pd.read_csv('C:\\Users\\Spencer\\Documents\\BUAD_5112(BB)\\CTBA\\M3\\ConstructionTimeSeriesDataV2.csv')
```

```python
fig, ax = plt.subplots()

ax.plot(construction['Month-Year'], construction['Private Construction'], alpha=.5)
ax.plot(construction['Month-Year'], construction['Public Construction'], alpha=.5)
plt.xlabel("Date", fontsize='14')
plt.ylabel("Spending in dollars", fontsize='14')
plt.suptitle("Private vs Public Construction Spending ", fontsize='16')
xtick = construction['Month-Year']
ax.set_xticks(xtick[::24])
ax.set_ylim(0,max(construction['Private Construction']+10000))
ax.legend(loc=1)
```

This graph visualizes the difference in spending between the public vs private sector. We can see that the private sector spends much more money on construction than the governemt. Additionally, we can infer that the crash of 2008 affected the construction spending in the private sector, but did not affect the public construction spending.
